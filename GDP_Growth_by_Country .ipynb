```python
# Dependencies and Setup
#import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#import time
import json
import requests
import pymysql
from sqlalchemy import create_engine
from sqlalchemy import Column, Integer, String, Float
pymysql.install_as_MySQLdb() 
```

```python
# Build query Url
url = "http://api.worldbank.org/v2/country/"
countries = ["ken","nga","gha","eth","rwa"] 
```

```python
# Create and initialize list 
# Setup dictionary to convert to df
country = []
indicator = []
year = []
percent = []

for i in countries:
    url_get = url + i +"/indicator/NY.GDP.MKTP.KD.ZG?format=json"  #/date=1980:2018
    response = requests.get(url_get)
    data = response.json()
    print(i)

    for item in data[1]:
        country.append(item['country']['value'])
        year.append(item['date'])
        indicator.append(item['indicator']['value'])
        percent.append(item['value'])
        print(item)
    with open(f'Resources/data/output_api.json', 'w') as outfile:json.dump({'country':country, 'indicator':indicator, 'year':year, 'percent':percent}, outfile) 
```

```python
df = pd.read_json("Resources/data/output_api.json")
```
